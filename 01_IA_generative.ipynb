# Leçon 1 — Qu’est-ce que l’IA générative ?

## Objectifs
- Comprendre discriminatif vs génératif
- Voir les familles de modèles (LLMs, diffusion, GAN, audio, multimodal)
- Identifier usages, risques, et limites

## Idées clés
- **Discriminatif** : p(y|x) — prédire une étiquette.
- **Génératif** : p(x) ou p(x|c) — générer du contenu (texte, image, audio...).
- **Méthodes** : auto-régression (GPT), auto-encodeurs, diffusion (Stable Diffusion), GANs.
- **Risques** : hallucinations, biais, droits d'auteur, attaques de prompt.

## Workflow générique
1. Choisir le modèle (hébergé ou local)
2. Définir le prompt / condition
3. Générer → filtrer → vérifier (guardrails)
4. Évaluer (qualité, coûts, latence)


```
# --- Génération de texte simple avec Transformers ---
from transformers import pipeline

# 1) Charger un pipeline de génération de texte.
#    Pour la démo CPU, un petit modèle comme 'gpt2' convient (anglais).
generator = pipeline('text-generation', model='gpt2')

# 2) Prompt concis et contextualisé (FR ou EN, selon le modèle).
prompt = "Explique en deux phrases ce qu'est l'IA générative en français :"

# 3) Générer un court texte.
outputs = generator(prompt, max_length=80, num_return_sequences=1)

# 4) Afficher le résultat.
print(outputs[0]['generated_text'])
```

> Remarque : Pour un modèle francophone léger, vous pouvez substituer un modèle compatible CPU disponible sur Hugging Face. Certains modèles nécessitent plus de RAM/VRAM.
